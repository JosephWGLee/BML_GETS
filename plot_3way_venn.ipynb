```python
import pandas as pd
from matplotlib_venn import venn3
import matplotlib.pyplot as plt

def plot_3way_venn(df1: pd.DataFrame, col1: str, df2: pd.DataFrame, col2: str, df3: pd.DataFrame, col3:str, labels: tuple, title: str, save_name: str):

    set1 = set(df1[col1])
    set2 = set(df2[col2])
    set3 = set(df3[col3])

    plt.figures(figsize(8,8))
    venn_diagrame = venn3([set1, set2, set3], set_labels=labels)
    plt.title(title, fontsize=12, color="black", ha="centre")
    plt.savefig(save_name, dpi=300)
    plt.show()
```
