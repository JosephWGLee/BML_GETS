```python
#Splitting of normalised counts data for batches (HPC)

import pandas as pd
import numpy as np

def split_dataset(df: pd.DataFrame, output_prefix: "normalised_counts_part"):
    
    split_dfs = np.array_split(df, 10)

    for i, split_df in enumerate(split_dfs):
        split_df.to_csv(f"{output_prefix}_{i+1}.csv", index=False)

print("Splitting complete!")
```
